We will work with breast cancer dataset to compare three different classifiers:

1. Logistic regression
2. Random forest classifier
3. K-nearest neighbors

Add the appropriate lines of code to answer the questions in each cell.

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer(as_frame=True)
```

```python
##################
### Question 1 ###
##################
# Import the test_train_split function 
from sklearn.model_selection import train_test_split
# Split the dataset so that 75% of the dataset is for training and 25% is for testing 
# Set the random state to 55
X = data["data"]
y = data["target"]
X_train,X_test,y_train,y_test = train_test_split(X.to_numpy(), y.to_numpy(), test_size = 0.25, random_state = 55)
# What is the value of 5th entry of y_test?
print(y_test[4])
# What is the value of row 3 and column 4 of X_train?
print(X_train[2,3])
# Remember that Python uses 0-indexing
```

```python
##################
### Question 2 ###
##################
# Scale the data to have 0 mean and unit variance
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
# What is the value of row 3 and column 4 of X_train?
print(X_train[2,3])
```

```python
##################
### Question 3 ###
##################
# Use the Scikit-Learn documentation and create 3 classifiers
from sklearn.linear_model import LogisticRegression
# 1. log_r is LogisticRegression, set the solver parameter to 'liblinear' and max_iter to 10
log_r = LogisticRegression(solver = 'liblinear', max_iter = 10, random_state = 55)
log_r.fit(X_train, y_train)
predictions_1 = log_r.predict(X_test)
print(predictions_1)
# 2. rnd_forest is RandomForestClassifier, set the n_estimators paramter to 5 and the max_depth to 3
from sklearn.ensemble import RandomForestClassifier
rnd_forest = RandomForestClassifier(n_estimators = 5, max_depth = 3, random_state = 55)
rnd_forest.fit(X_train, y_train)
predictions_2 = rnd_forest.predict(X_test)
print(predictions_2)
# 3. k_near is K-Nearest Neighbor classifier, set the n_neighbors parameter to 4
from sklearn.neighbors import KNeighborsClassifier
k_near = KNeighborsClassifier(n_neighbors = 4)
k_near.fit(X_train, y_train)
predictions_3 = k_near.predict(X_test)
print(predictions_3)
```

```python
##################
### Question 4 ###
##################
# Report the accuracy of all three models
score_1 = log_r.score(X_test, y_test)
print("The accuracy of the logistic regression model is: ", score_1)
score_2 = rnd_forest.score(X_test, y_test)
print("The accuracy of the random forest classifier model is: ", score_2)
score_3 = k_near.score(X_test, y_test)
print("The accuracy of the k-nearest neighbor classifier model is: ", score_3)
# Which model is the most accurate?
print("K-nearest Neighbor Classifier has the highest score.")
```

```python
##################
### Question 5 ###
##################
# Compute the precision for each model
from sklearn import metrics
precision_1 = metrics.precision_score(y_test, predictions_1)
print("Precision score of logistic regression model: ", precision_1)
precision_2 = metrics.precision_score(y_test, predictions_2)
print("Precision score of random forest classifier model: ", precision_2)
precision_3 = metrics.precision_score(y_test, predictions_3)
print("Precision score of k-nearest neighbor classifier model: ", precision_3)
# Which model(s) has the best precision?
print("K-nearest Neighbor Classifier has the highest score.")
```

```python
##################
### Question 6 ###
##################
# Compute the recall for each model
recall_1 = metrics.recall_score(y_test, predictions_1)
print("Recall score of logistic regression model: ", recall_1)
recall_2 = metrics.recall_score(y_test, predictions_2)
print("Recall score of random forest classifier model: ", recall_2)
recall_3 = metrics.recall_score(y_test, predictions_3)
print("Recall score of k-nearest neighbor classifier model: ", recall_3)
# Which model(s) has the best recall?
print("Logistic Regression has the highest score.")
```

```python
##################
### Question 7 ###
##################
# Plot the confusion matrix for the model with the 2nd- highest accuracy.
# Ensure the title contains the accuracy and score of the model.
import matplotlib.pyplot as plt
import seaborn as sns
cm = metrics.confusion_matrix(y_test, predictions_1)
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score_1)
plt.title(all_sample_title, size = 15);
```

```python
##################
### Question 8 ###
##################
# Which model would you pick to predict whether a patient had breast cancer or not?
print("I choose K-nearest Neighbor Classifier Model.")
# Why would you choose this model?
print("Because it has the highest accuracy.")
```

```python
##################
### Question 10###
##################
# Plot the ROC curve for the model selected in Question 8.
lr_probs = k_near.predict_proba(X_test)
lr_preds = lr_probs[:,1]
fpr, tpr, threshold = metrics.roc_curve(y_test, lr_preds)
print(threshold)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([-0.01, 1])
plt.ylim([0, 1.01])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
